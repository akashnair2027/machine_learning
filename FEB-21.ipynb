```python
tele_data = df.copy()
```

```python
tele_data.TotalCharges = pd.to_numeric(tele_data.TotalCharges,errors = 'coerce')
```

```python
tele_data.isnull().sum()
```

```python
tele_data.dtypes
```

```python
tele_data.loc[tele_data["TotalCharges"].isnull()==True]
```

```python
tele_data.isna().sum()
```

```python
tele_data.dropna(how = "any",inplace = True)
```

```python
tele_data.shape
```

```python
tele_data.head()
```

```python
labels = [ "{0} - {1}".format(i,i+11) for i in range (1,72,12)]
tele_data['tenure_group'] =pd.cut(tele_data.tenure,range(1,80,12),right = False,labels = labels)
```

```python
tele_data['tenure_group'].value_counts()
```

```python
tele_data.drop(columns = ["customerID","tenure"],axis =1,inplace =True)
```

```python
tele_data.head()
```

```python
#univeariate analysis
for i,predicator in enumerate(tele_data.drop(columns = ["Churn","TotalCharges","MonthlyCharges"])):
    plt.figure(i)
    sns.countplot(data = tele_data,x = predicator,hue = "Churn")    
```

```python
tele_data_dummies=pd.get_dummies(tele_data)
tele_data_dummies
```

```python
sns.lmplot(data = tele_data_dummies,x = "MonthlyCharges",y = 'TotalCharges',fit_reg = False)
```

```python
tele_data['Churn']=np.where(tele_data.Churn=='Yes',1 ,0)
```

```python
tele_data.head()
```

```python
tele_data_dummies =pd.get_dummies(tele_data) 
```

```python
tele_data_dummies.head()
```

```python
mth = sns.kdeplot(tele_data_dummies.MonthlyCharges[(tele_data_dummies["Churn"]==0)],color = "Red",shade= True)
mth = sns.kdeplot(tele_data_dummies.MonthlyCharges[(tele_data_dummies["Churn"]==1)],color = "blue",shade=True)

mth.legend(["No Churn","Churn"],loc = "upper right")
mth.set_ylabel("Density")
mth.set_xlabel("Monthly Charges")
mth.set_title("Monthly Charges by churn")
```

```python
plt.figure(figsize = (20,8))
tele_data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind = 'bar')
```
