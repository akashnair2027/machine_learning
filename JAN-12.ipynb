```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of sample
n =[2,20,30,100,1000,10000]
    # list for sample means
    
sm =[]
# Generating 2,30,100,1000,10000 random numbers from -40 to 80
# taking their mean appending it to list means


for i in n :
    np.random.seed(1)
    x =[np.mean(np.random.randint(-40,40,i))for j in range(10000) ]
    sm.append(x)
k =0

fig,ax = plt.subplots(2,3,figsize = (10,10))
for i in range(0,2):
    for j in range(0,3):
        ax[i,j].hist(sm[k],10,density = True)
        ax[i,j].set_title(label = n[k])
        k = k+1
plt.show()
```
