```python
x= df1["day"]
y = df1["total_bill"]
plt.xlabel("DAY")
plt.bar(x,y,color = "Green",edgecolor = "blue",linewidth = 1)
plt.show()
```

```python
x= df1["total_bill"]
plt.hist(x,bins = 25,color = "red",edgecolor = "blue",linestyle = "--",alpha = 0.5)
plt.title("Tips dataset")
plt.xlabel("Frequency")
plt.ylabel("Total bill")
plt.show()
```

```python
#scatter plot 
y = df1["total_bill"] 
x = df1["tip"]

#plotting the data
plt.scatter(y,x)

#add tittle
plt.title("Tips dataset")

#label on x-Axis
plt.xlabel("Tips")

#label on y axis
plt.ylabel("total_bill")
```

```python
df1.plot(kind = "scatter",x = "day",y = "total_bill")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
#X-axis points
y = x*2
#y -axis points
plt.plot(x,y)
#plot the chart
plt.scatter(x,y,c = "pink",linewidth = 2,
           marker = "s",#""^"",
            edgecolor ="green",s = 50)
```

```python
cars =["BMW","JAGUAR","TESLA","AUDI","FORD"]
Data = [23,43,13,35,12]
e = [0,0.5,0,0,0]
col= ("orange","cyan","yellow","green","Blue")
plt.pie(Data,labels=cars,explode = e,autopct="%1.2f%%",colors = col,shadow = True)
plt.savefig("Image.jpg")
plt.show()
```
