As a rule of thumb, about 68% of the data lies within 1 standard deviation of the mean, 95% lies within 2 standard deviations and 99.7% lies within 3 standard deviations

Eg: Heights of adults in a town and the data follows a normal distribution, we have a sufficient sample size with mean equals 5.3 and the standard deviation is 1.

    The norm.pdf()class method requires loc and scale ( mean and sd)

    Now, if we were asked to pick one person randomly from this distribution, then what is the probability that the height of the person will be smaller than 4.5 ft. ?
The area under the curve is nothing but just the Integration of the density function with limits

```python
import numpy as np
import matplotlib.pyplot as plt

#creating a series of data in range of 1-50
x = np.linspace(1,10,20)
def nr(x,mean,sd):
        dp = (np.pi*sd)*np.exp(-0.5*((x-mean)/sd)**2)
        return dp
    
```

```python
mean = np.mean(x)
sd = np.std(x)
    
pdf = nr(x,mean,sd)
    
plt.plot(x,pdf,color = "red")
plt.xlabel = ("Height")
plt.ylabel = ("Probability")
```

the area under the curve is nothing but just the intergration of the density functiom with limits equal = to 4.5
