```python
import matplotlib.pyplot as plt
%matplotlib inline
# toshow plots that are embedded to jupytter notebook
# the simply way to create a graph ois using the plot() method
import numpy as np
x = np.linspace(0,50,30)
y = x*np.linspace(10,100,30)
```

```python
plt.plot(x,y,color = "green",linewidth = 1,marker ="o",markersize = 3,linestyle ="--")
#plt.plot()

plt.title("LineChart",fontsize =10, color ="Red")
plt.xlabel("Quality")
plt.ylabel("Sales")
plt.legend("[sales]")
plt.show()
```

```python
x = [10,30,50,70]
y = [20,40,60,80]
#creating figure object
plt.figure()

#adding first subplot 
plt.subplot(121)
plt.plot(x,y)

#adding second subplot
plt.subplot(122)
plt.plot(y,x)
```
